```python
ricePrice = new_df['DZDPrice'].where(new_df['commodity']=='Rice')
breadPrice = new_df['DZDPrice'].where(new_df['commodity']=='Bread')
sugarPrice = new_df['DZDPrice'].where(new_df['commodity']=='Sugar')

```

```python
ricePrice = ricePrice.dropna()
breadPrice = breadPrice.dropna()
sugarPrice = sugarPrice.dropna()
```

```python
plt.figure(figsize=(10,5))

plt.plot(df['date'].where(new_df['commodity']=='Rice').dropna(),ricePrice,c='blue', label='Rice Price')
plt.plot(df['date'].where(new_df['commodity']=='Bread').dropna(),breadPrice,c='red',label='Bread Price')
plt.plot(df['date'].where(new_df['commodity']=='Sugar').dropna(),sugarPrice,c='green',label='Sugar Price')
plt.legend()
plt.show()
```

```python
new_df1 = new_df.groupby(['category', 'commodity','Year']).std()
new_df1
```

```python
test_df = new_df.copy()
eggs = new_df['commodity'].where(new_df['commodity'] == 'Eggs').dropna()
hydra = new_df['City'].where(new_df['City'] == 'Hydra').dropna()
test_df['commodity'] = eggs
test_df = test_df.dropna()
test_df['City'] = hydra
test_df = test_df.dropna()
eggs_price = test_df.groupby(['commodity','Month']).std()
eggs_price
```

```python
tindouf_df = new_df.copy()
tea = new_df['commodity'].where(new_df['commodity'] == 'Tea (sahm)').dropna()
tindouf = new_df['State'].where(new_df['State'] == 'Tindouf').dropna()
tindouf_df['commodity'] = tea
tindouf_df['State'] = tindouf
tindouf_df.dropna()
tea_Price = tindouf_df.groupby(['commodity','State','Month']).std()
tea_Price
```

```python
year_price = new_df.copy()
Y2015 = new_df['Year'].where(new_df['Year'] == 2015).dropna()
tindouf = new_df['State'].where(new_df['State'] == 'Tindouf').dropna()
year_price['Year'] = Y2015
tindouf_df['State'] = tindouf
year_price = year_price.dropna()
year_price = year_price.sort_values(by='DZDPrice', ascending=False)
year_price = year_price.drop(['date', 'State', 'City', 'latitude', 'longitude', 'unit','Month'], axis=1)

prices=year_price.groupby(['commodity'])['DZDPrice'].max().sort_values()
prices
```
